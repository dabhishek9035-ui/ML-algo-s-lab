```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("tumor.csv")
df.head()
```

```python
X = df[['radius', 'texture', 'perimeter', 'area']].values
y=df["diagnosis"].values
```

```python
def GDA(X, y):
    
    m, n = X.shape
    phi = np.sum(y) / m
    mu1 = np.sum(y.reshape(-1, 1) * X, axis=0) / np.sum(y)
    mu0 = np.sum((1 - y).reshape(-1, 1) * X, axis=0) / np.sum(1 - y)
    mu1 = np.sum(y.reshape(-1, 1) * X, axis=0) / np.sum(y)
    mu0 = np.sum((1 - y).reshape(-1, 1) * X, axis=0) / np.sum(1 - y)
    Sigma = np.zeros((n, n))
    for i in range(m):
        if y[i] == 1:
            k = X[i] - mu1
        else:
            k = X[i] - mu0
        Sigma += np.outer(k, k)
    Sigma = Sigma / m
    return phi, mu0, mu1, Sigma,n
```

```python
def predict(x, phi, mu0, mu1, Sigma,n):
    
    P1 = phi * (np.exp(-0.5 * ((x-mu1).T @ np.linalg.inv(Sigma) @ (x-mu1))) /
                (((2*np.pi)**(n/2)) * (np.linalg.det(Sigma)**0.5)))
    P0 = (1 - phi) * (np.exp(-0.5 * ((x-mu0).T @ np.linalg.inv(Sigma) @ (x-mu0))) /
                        (((2*np.pi)**(n/2)) * (np.linalg.det(Sigma)**0.5)))
    if P1 > P0:
        return f"the person has a tumor and the probability of having a tumor is {P1/(P1+P0)*100}%"
    else:
        return f"the person does not have a tumor and the probability of not having a tumor is {P0/(P1+P0)*100}%"
```

```python
phi,mu0,mu1,sigma,n=GDA(X,y)
```

```python
r=float(input("enter radius of tumor: "))
t=float(input("enter texture of tumor: "))
perimeter=float(input("enter perimeter of tumor: "))
area=float(input("enter area of tumor: "))
x= [r,t,perimeter,area]
prediction=predict(x,phi,mu0,mu1,sigma,n)
print(prediction)
```
