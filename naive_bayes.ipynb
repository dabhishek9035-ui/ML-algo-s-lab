```python
# The dataset is not actually Bernoulli's, we are using this algo assuming it to be one.(better to use multinomial naive bayes for this dataset)
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("spam.csv")
df.head()
```

```python
X = df.drop("label", axis=1)
y = df['label']
```

```python
def naive_bayes(X, y): 
    n=len(y)
    classes=np.unique(y)
    p0=np.zeros(20)
    p1=np.zeros(20)
    for i in range(20):
        c0=0
        c1=0
        for j in range(n):
            if y.iloc[j]==classes[0]:
                if X.iloc[j, i] !=0:
                    c0+=1
            else:
                if X.iloc[j, i] !=0:
                    c1+=1
        p0[i]=c0/np.sum(y == classes[0])
        p1[i]=c1/np.sum(y == classes[1])
    prior0 = np.sum(y == classes[0]) / n
    prior1 = np.sum(y == classes[1]) / n
    return prior0, prior1, p0, p1
```

```python
prior0, prior1, p0, p1=naive_bayes(X, y)
```

```python
def predict(X, p0, p1, prior0, prior1):
    spam_prob = prior1
    not_spam_prob = prior0
    for i in range(20):
        if X[i] != 0:
            spam_prob *= p1[i]
            not_spam_prob *= p0[i]
    if spam_prob > not_spam_prob:
        return f"the email is likely spam"
    else:
        return f"the email is not spam"
```

```python
def parser(message):
    names=df.columns.tolist()
    counts={}
    for i in message.lower().split():
        if i in names:
            if i in counts:
                counts[i]+=1
            else:
                counts[i]=1
    ordered_dict = {key: counts.get(key, 0) for key in names}
    lst=list(ordered_dict.values())
    return lst
```

```python
message=input("enter ur message: ")
z=parser(message)
result=predict(z, p0, p1, prior0, prior1)
print(result)
```
