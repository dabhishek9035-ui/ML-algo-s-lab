```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("spam.csv")
df.head()
```

```python
X=df[::-1]
y=df['label']
```

```python
def naive_bayes(X, y): #using guassian expression, assuming var=constant to simplify code and calculations
    n=len(y)
    classes=np.unique(y)
    mu0=np.mean(X[y==classes[0]], axis=0)
    mu1=np.mean(X[y==classes[1]], axis=0)
    return mu0, mu1,classes
```

```python
mu0, mu1,classes=naive_bayes(X, y)
```

```python
def predict(X, mu0, mu1):
    prior0 = np.sum(y == classes[0]) / len(y)
    prior1 = np.sum(y == classes[1]) / len(y)
    p0 = np.prod((1/np.sqrt(2*np.pi))*np.exp(-(X-mu0)**2/2))*prior0
    p1 = np.prod((1/np.sqrt(2*np.pi))*np.exp(-(X-mu1)**2/2))*prior1
    if p0>p1:
        return f"the email is likely spam"
    else:
        return f"the email is not spam"
```

```python
def parser(message):
    names=df.columns.tolist()
    counts={}
    for i in message.lower().split():
        if i in names:
            if i in counts:
                counts[i]+=1
            else:
                counts[i]=1
    ordered_dict = {key: counts.get(key, 0) for key in names}
    lst=list(ordered_dict.values())
    return lst
```

```python
message=input("enter ur message: ")
z=parser(message)
result=predict(z, mu0, mu1)
print(result)
```
