```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Salary_dataset.csv")
df.head()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
X=df['YearsExperience']
y=df['Salary']
```

```python
def linear_regression(X, y):
    X = np.array(X).reshape(-1, 1)                  #reshaping the X and y values to be in the correct format for matrix operations
    y = np.array(y).reshape(-1, 1)
    X = np.c_[np.ones((X.shape[0],1)), X]           #Adding a column of ones to the X values to account for the bias term in the linear regression
    X_t=X.T
    theta_best=(np.linalg.inv(X_t@X))@(X_t@y)       #Normal equation method to find the best theta values
    return theta_best
```

```python
a=linear_regression(X,y)
print(a)
```

```python
inp=float(input("enter the years of experience: ")) #predicting the salary based on the years of experience entered by the user
result=a[0][0]+a[1][0]*inp
print(f"the estimated salary is {result}")
```
