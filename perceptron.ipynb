```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('student_pass.csv')
df.head()
```

```python
X1=df['hours_studied']
X2=df['attendance_pct']
X=pd.concat([X1,X2],axis=1)
y=df['passed']
```

```python
def step_function(z):
    return 1 if z>=0 else 0
```

```python
def perceptron(X,y):
    X = np.hstack((np.ones((X.shape[0], 1)), X))
    theta=np.zeros(X.shape[1])

    for i in range(100):
        for j in range(X.shape[0]):
            g=step_function(np.dot(theta,X[j]))
            theta=theta+(y[j]-g)*X[j]
    return theta
```

```python
a=perceptron(X,y) #a=[-46. 189.  -2.] The values are not accurate, due to high difference in the values of hours studied and attendance percentage. The model is not able to converge properly. We need to scale the values to bring them to a similar range. But since this is a simple example used for practice, we will not scale the values here.
print(a)
```

```python
hrs=int(input("Enter the number of hours studied: "))
att=int(input("Enter the attendance percentage: "))
x=[1,hrs,att]
out=step_function(np.dot(a,x))
if out==1:
    print("The student is likely to pass the exam.")
else:
    print("The student is likely to fail the exam.")
```
