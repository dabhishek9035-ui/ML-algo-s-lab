```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("shapes.csv")
df.head()
```

```python
X1=df['sides']
X2=df['compactness']
X=np.concatenate((X1,X2))
y=df["shape"]
X = X.reshape(-1, 1)
```

```python
def labelEncoder(y):
    classes = np.unique(y)
    class_to_index = {cls: idx for idx, cls in enumerate(classes)}
    y_encoded = np.array([class_to_index[label] for label in y])
    return y_encoded
```

```python
def softmax(X,y):
    X = np.hstack((np.ones((X.shape[0], 1)), X))
    theta=np.zeros((X.shape[1], len(np.unique(y))))
    for i in range(100):
        for j in range(len(y)):
            z=X[j]@theta
            z=z-np.max(z)
            p=np.exp(z)/np.sum(np.exp(z))
            y_target = np.zeros(theta.shape[1])  
            y_target[y[j]] = 1.0
            J = np.outer(X[j], p - y_target)
            theta=theta-J
    return theta
```

```python
y=labelEncoder(y)
a=softmax(X,y)
print(a)
```

```python
m=int(input("Enter the number of sides: "))
z_n=float(input("Enter the compactness: "))
x=np.array([[m,z_n]])
z = x @ a
p=np.exp(z)/np.sum(np.exp(z))
idx=np.argmax(p)
classes=["Triangle", "Circle",  "rectangle" ]
shape=classes[idx]
print(f"The predicted shape is: {shape}")
```
