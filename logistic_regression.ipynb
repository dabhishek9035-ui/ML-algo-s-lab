```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('placements.csv')
df.head()
```

```python
X1=df['iq']
X2=df['cgpa']
X=pd.concat([X1,X2],axis=1)
y=df['placements']
```

```python
def logistic_regression(X,y):
    X = np.hstack((np.ones((X.shape[0], 1)), X))
    theta=np.zeros(X.shape[1])  
    alpha=0.01
    num_iterations=len(y)*1000
    h=np.zeros(num_iterations)
    for i in range(num_iterations):
        for j in range(len(X)):
            z=X[j]@theta
            h[j]=1/(1+np.exp(-z))
            theta = theta + (alpha * (y[j] - h[j]) * X[j])
    return theta
```

```python
a=logistic_regression(X,y)
print(a)
```

```python
iq= int(input("Enter the IQ of the student: "))
cgpa= float(input("Enter the CGPA of the student: "))
x=np.array([[1, iq,cgpa]])
z=x@a
h=1/(1+np.exp(-z))
if h>=0.5:
    print("The student is likely to be placed.")
else:
    print("The student is unlikely to be placed.")
```
