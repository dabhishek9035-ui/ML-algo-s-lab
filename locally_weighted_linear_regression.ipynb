```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("car_fuel_eff.csv")
df.head()
```

```python
X=df['speed_kmph']
y=df['fuel_efficiency_kmpl']
```

```python
def locally_weighted_linear_regression(X, y,x):
    n=len(X)
    X = np.array(X).reshape(-1, 1)                 
    y = np.array(y).reshape(-1, 1)
    theta=np.zeros(X.shape[0])
    w = np.zeros(X.shape[0])
    for i in range(n):
        diff = X[i,0] - x
        w[i] = np.exp(-(diff**2) / (2))
    X = np.c_[np.ones((X.shape[0],1)), X]  
    X_t=X.T    
    W = np.diag(w)
    theta_best = np.linalg.inv(X_t @ W @ X) @ (X_t @ W @ y)    
    return theta_best
    
```

```python
inp=float(input("Enter the speed of the car in kmph: "))
a=locally_weighted_linear_regression(X, y,inp)
print(a)  #the value of a varies acc to input 
result=a[0]+a[1]*inp
print("The fuel efficiency of the car at speed",inp,"kmph is",result)
```
