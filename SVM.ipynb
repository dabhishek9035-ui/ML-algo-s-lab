```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('exams.csv')
df.head()
```

```python
X=df[['study_hours','prev_score']]
y=df['result']
```

```python
def SVM(X,y,C=1.0):
    m,n=X.shape
    alpha=np.zeros(m)
    alpha1=np.zeros(m)
    b=0
    for epoch in range(10):
        for i in range(m):
            for j in range(m):
                if i==j:
                    continue
                alpha1 = alpha.copy()
                Kii = np.dot(X[i], X[i])
                Kij = np.dot(X[i], X[j])
                Kjj = np.dot(X[j], X[j])
                Ei = np.sum(alpha * y * (X @ X[i])) + b - y[i]
                Ej = np.sum(alpha * y * (X @ X[j])) + b - y[j]
                eeta=Kii+Kjj-2*Kij
                if eeta<0:
                    continue
                if y[i]!=y[j]:
                    L=np.maximum(0,alpha[j]-alpha[i])
                    H=np.minimum(C,C+alpha[j]-alpha[i])
                else:
                    L=np.maximum(0,alpha[j]+alpha[i]-C)
                    H=np.minimum(C,alpha[i]+alpha[j])
                alpha1[j]=alpha[j]+y[j]*(Ei-Ej)/eeta
                if alpha1[j] > H:
                    alpha1[j] = H
                elif alpha1[j] < L:
                    alpha1[j] = L
                alpha1[i]=alpha[i]+y[i]*y[j]*(alpha[j]-alpha1[j])

                b1 = (b-Ei- y[i] * (alpha1[i] - alpha[i]) * Kii- y[j] * (alpha1[j] - alpha[j]) * Kij)
                b2 = (b-Ej-y[i]*(alpha1[i]-alpha[i])*Kij-y[j]*(alpha1[j]-alpha[j]) * Kjj)
                if 0 < alpha1[i] < C:
                    b = b1
                elif 0 < alpha1[j] < C:
                    b = b2
                else:
                    b = (b1 + b2) / 2
                alpha = alpha1.copy()
        w = (alpha*y)@X
    return w,b,alpha
```

```python
def predict(X,w,b):
    prediction=np.dot(X,w)+b
    if prediction>=0:
        return 1
    else:
        return -1
```

```python
w, b, alpha = SVM(X.to_numpy(), y.to_numpy())
```
